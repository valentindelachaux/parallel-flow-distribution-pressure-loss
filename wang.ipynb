```python
from math import sqrt, cos, sin, acos, asin, exp
```

```python
# Define the variables
beta_i = 0.5 #pressure recovery factor in the inlet manifold
beta_e = 0.5 #pressure recovery factor in the exhaust manifold
zeta = 1.5 # average total head loss coefficient for channel flow
D_i = 2 # diameter of inlet manifold
D_e = 2 # diameter of exhaust manifold
Fi = 5 # cross sectionnal area of inlet manifold
Fe = 5 # cross sectionnal area of exhaust manifold
Fc = 10 # cross sectionnal area of channels
n = 2 # number of channels
L = 1.5 # length of header (m)

v = 3
Re_i = 200


# Calculate A1
A1 = ((2 - beta_i) / (3 * zeta)) * ((Fc * n) / Fi)**2

# Calculate A2
A2 = ((2 - beta_e) / (3 * zeta)) * ((Fc * n) / Fe)**2

# Calculate B1
B1 = -(L / (4 * zeta)) * (Fi / D_i) * ((Fc * n) / Fi)**2

# Calculate epsilon
epsilon = -(Fe * L) / (4 * zeta * D_e) * ((Fc * n) / Fe)**2

# Calculate M
M = (Fi * D_i) / (v * (Re_i * Fi))

# Calculate the equations
Q = A1 - (1 - M) * A2
R = B1 = (1 - 2*M)*epsilon
```

```python
# Calculate B
B = (R + sqrt(Q**3 + R**2))**(1/3) + (R - sqrt(Q**3 + R**2))**(1/3)

# Calculate J
J = (R + sqrt(Q**3 + R**2))**(1/3) - (R - sqrt(Q**3 + R**2))**(1/3)

# Calculate r
r = B

# Calculate r1
r1 = -0.5 * B + 0.5j * sqrt(3) * J

# Calculate r2
r2 = -0.5 * B - 0.5j * sqrt(3) * J
```
