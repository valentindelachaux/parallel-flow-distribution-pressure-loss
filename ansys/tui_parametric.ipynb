```python
import os
from io import StringIO

import numpy as np
import pandas as pd
import openpyxl

import matplotlib.pyplot as plt
import plotly.graph_objects as go

import jou_gen as jg

import ansys.fluent.core as pyfluent
```

```python
folder_path = r'D:\ANSYS Fluent Projects\Hydraulics\V4.5\3MPE'
folder_path_param = r'D:\ANSYS Fluent Projects\Hydraulics\V4.5\3MPE\parametric1'

inputs = pd.read_excel(folder_path+'\\3MPE_tui-parametric.xlsx')

solver_path = folder_path_param + '\\' + 'server_info-24316.txt'
solver = pyfluent.connect_to_fluent(server_info_filepath=solver_path)
tui = solver.tui
```

```python
inputs
```

```python
for row, value in inputs.iterrows():

    if row>=5:

        print(value['Name'])

        jg.change_named_expression(tui, 'mdot_distrib_inlet', value['mdot_distrib_inlet'], 'kg/s')
        jg.change_named_expression(tui, 'mdot_distrib_outlet', value['mdot_distrib_outlet'], 'kg/s')
        jg.change_named_expression(tui, 'mdot_coll_inlet', value['mdot_coll_inlet'], 'kg/s')

        solver.solution.initialization.hybrid_initialize()
            
        solver.solution.run_calculation.iterate(number_of_iterations=value['nb_it'])

        jg.write_report(tui, 'mdot', folder_path_param, value['Name']+f'_mdot_report')
        jg.write_report(tui, 'sp', folder_path_param, value['Name']+f'_sp_report')
        jg.write_plot_xy(tui, 'line-distrib-sp', folder_path_param, value['Name']+'_line-distrib-sp')
        jg.write_plot_xy(tui, 'line-coll-sp', folder_path_param, value['Name']+'_line-coll-sp')
        jg.write_residuals_file(tui, folder_path_param, value['Name']+'_residuals')
        jg.write_time(tui, folder_path_param, value['Name']+'_time')
        jg.write_data(tui, folder_path_param, value['Name']+'_data')
```
